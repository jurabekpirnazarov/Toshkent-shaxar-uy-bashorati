```python
def predict_price(location,rooms,size,level,max_level):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = rooms
    x[1] = size
    x[2] = level
    x[3] = max_level
    if loc_index >= 0:
        x[loc_index] = 1

    return LR_model.predict([x])[0]
```
